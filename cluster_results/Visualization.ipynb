```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("comparison.xlsx")
```

```python
df
```

```python
plt.figure(figsize=(5,5))
plt.scatter(df["ECP"].values, df["ECA"].values, c = df.index)
plt.title("Single Cell 3 and 5 lines")
plt.xlabel("Entropy of cluster purity")
plt.ylabel("Entropy of cluster accuracy")

for i, (a,b,c) in enumerate(zip(df["ECP"].values, df["ECA"].values, df["methods"])):
    plt.text(a, b+(0.02-i*0.01)*(-1)**i, '%s' % c, ha='center', va= 'bottom',fontsize=9)
plt.show()

plt.figure(figsize=(15,5))
plt.bar(df["methods"], df["ARI"])
plt.title("Single Cell 3 and 5 lines")
plt.xlabel("methods")
plt.ylabel("Entropy of cluster accuracy")
```
